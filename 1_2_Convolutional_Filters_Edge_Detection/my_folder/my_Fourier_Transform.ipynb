```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
image_stripes = cv2.imread("..\\images\\stripes.jpg")
image_stripes = cv2.cvtColor(image_stripes, cv2.COLOR_BGR2RGB)
# cv2.imshow("Stripes", image_stripes)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
image_solid = cv2.imread("..\\images\\pink_solid.jpg")
image_solid = cv2.cvtColor(image_solid, cv2.COLOR_BGR2RGB)
# cv2.imshow("Stripes", image_solid)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax[0].imshow(image_stripes)
ax[1].imshow(image_solid)
```

Change color to gray scale

```python
gray_stripes = cv2.cvtColor(image_stripes, cv2.COLOR_RGB2GRAY)
gray_solid = cv2.cvtColor(image_solid, cv2.COLOR_RGB2GRAY)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax[0].imshow(gray_stripes, cmap='gray')
ax[1].imshow(gray_solid, cmap='gray')
```

```python
norm_stripes = gray_stripes / 255.0
norm_solid = gray_solid / 255.0

# Perfom FFT and create transformed scaled image

def ft_image(image):
    f = np.fft.fft2(image)
    fshift = np.fft.fftshift(f)
    frequency_tx = 20 * np.log(np.abs(fshift))

    return frequency_tx
```

```python
f_stripes = ft_image(norm_stripes)
f_solids = ft_image(norm_solid)

fig, ax = plt.subplots(nrows=2 , ncols=2, figsize=(10, 10))
ax[0][0].imshow(image_stripes, cmap='gray')
ax[0][0].set_title("Original stripes")
ax[0][1].imshow(f_stripes, cmap='gray')
ax[0][1].set_title("Fourier stripes")

ax[1][0].imshow(image_solid, cmap='gray')
ax[1][0].set_title("Original solids")
ax[1][1].imshow(f_solids, cmap='gray')
ax[1][1].set_title("Fourier solid")
```

Look at the real world image

```python
image = cv2.imread("..\\images\\birds.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

norm_image = gray / 255.0

f_image = ft_image(norm_image)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax[0].imshow(image)
ax[1].imshow(f_image, cmap='gray')
```
