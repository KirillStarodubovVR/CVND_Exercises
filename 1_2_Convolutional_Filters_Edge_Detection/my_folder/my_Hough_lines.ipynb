```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread("..\\images\\phone.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap="gray")
```

```python
sigma = 0.33
v = np.median(gray)

lower = int(max(0, (1.0 - sigma)*v))
upper = int(min(255, (1.0 + sigma)*v))
edge = cv2.Canny(gray, lower, upper)

plt.imshow(edge, cmap="gray")
```

```python
edge2 = cv2.Canny(gray, 100, 230)
plt.imshow(edge2, cmap="gray")
```

```python
rho = 3
theta = np.pi / 180
threshold = 100
min_line_length = 100
max_line_gap = 5

line_image = np.copy(image)
lines = cv2.HoughLinesP(edge, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)

for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image, (x1, y1), (x2, y2), (255, 0, 0), 5)

plt.imshow(line_image)
```

```python
rho = 1
theta = np.pi / 180
threshold = 70
min_line_length = 110
max_line_gap = 6

line_image = np.copy(image)
lines = cv2.HoughLinesP(edge2, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)

for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image, (x1, y1), (x2, y2), (255, 0, 0), 5)

plt.imshow(line_image)
```
