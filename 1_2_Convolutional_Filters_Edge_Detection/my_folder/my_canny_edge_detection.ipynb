```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('..\\images\\brain_MR.jpg')
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (5,5), 0)
plt.imshow(blurred, cmap="gray")
```

```python
wide = cv2.Canny(blurred, 30, 100)
tight = cv2.Canny(blurred, 200, 240)

f, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,6))
ax[0].set_title("wide")
ax[0].imshow(wide, cmap="gray")

ax[1].set_title("tight")
ax[1].imshow(tight, cmap="gray")
```

```python
image = cv2.imread('..\\images\\sunflower.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
blurred = cv2.GaussianBlur(gray, (7, 7), 0)

f, ax = plt.subplots(nrows=1, ncols=3, figsize=(12,6))
ax[0].set_title("original")
ax[0].imshow(image, cmap="gray")

ax[1].set_title("gray")
ax[1].imshow(gray, cmap="gray")

ax[2].set_title("blurred")
ax[2].imshow(blurred, cmap="gray")
```

```python
wide = cv2.Canny(blurred, 10, 200)
mid = cv2.Canny(blurred, 30, 150)
tight = cv2.Canny(blurred, 200, 240)

f, ax = plt.subplots(nrows=1, ncols=3, figsize=(12,6))
ax[0].set_title("wide")
ax[0].imshow(wide, cmap="gray")

ax[1].set_title("mid")
ax[1].imshow(mid, cmap="gray")

ax[2].set_title("tight")
ax[2].imshow(tight, cmap="gray")
```

```python
# apply automatic Canny edge detection using the computed median
v = np.median(blurred)
sigma = 0.33

lower = int(max(0, (1.0 - sigma) * v))
upper = int(min(255, (1.0 + sigma) * v))
edged = cv2.Canny(blurred, lower, upper)

f, ax = plt.subplots(nrows=1, ncols=4, figsize=(12,6))
ax[0].set_title("wide")
ax[0].imshow(wide, cmap="gray")

ax[1].set_title("mid")
ax[1].imshow(mid, cmap="gray")

ax[2].set_title("tight")
ax[2].imshow(tight, cmap="gray")

ax[3].set_title("auto")
ax[3].imshow(edged, cmap="gray")
```
