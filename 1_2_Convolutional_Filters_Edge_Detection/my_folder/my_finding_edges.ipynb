```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2
import numpy as np
```

```python
image = mpimg.imread("..\\images\\round_farms.jpg")
plt.imshow(image)
```

```python
gray = cv2.cvtColor(np.copy(image), cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
sobelX = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])

sobelX_rev = np.array([[1, 0, -1], 
                       [2, 0, -2], 
                       [1, 0, -1]])

sobelY = np.array([[-1, -2, -1], 
                   [0, 0, 0], 
                   [1, 2, 1]])

sobelY_rev = np.array([[1, 2, 1], 
                       [0, 0, 0], 
                       [-1, -2, -1]])

sharpen = np.array(([[0, -1, 0], 
                     [-1, 5, -1], 
                     [0, -1, 0]]), dtype="int")

laplacian = np.array(([[0, 1, 0], 
                       [1, -4, 1], 
                       [0, 1, 0]]), dtype="int")

laplacian2 = np.array(([[-1, -1, -1], 
                        [-1, 8, -1], 
                        [-1, -1, -1]]), dtype="int")

smallBlur = np.ones(shape=(7,7), dtype="uint8") / (7 * 7)
```

```python
fig, ax = plt.subplots(nrows=5, ncols=2, figsize=(10, 20))

ax[0][0].set_title("Original image")
ax[0][0].imshow(image)

ax[0][1].set_title("Gray image")
ax[0][1].imshow(gray, cmap="gray")

filtered_image = cv2.filter2D(gray, -1, sobelX)
ax[1][0].set_title("SobelX image")
ax[1][0].imshow(filtered_image, cmap="gray")

filtered_image = cv2.filter2D(gray, -1, sobelX_rev)
ax[1][1].set_title("SobelX reverse image")
ax[1][1].imshow(filtered_image, cmap="gray")

filtered_image = cv2.filter2D(gray, -1, sobelY)
ax[2][0].set_title("SobelY image")
ax[2][0].imshow(filtered_image, cmap="gray")

filtered_image = cv2.filter2D(gray, -1, sobelY_rev)
ax[2][1].set_title("SobelY reverse image")
ax[2][1].imshow(filtered_image, cmap="gray")

filtered_image = cv2.filter2D(image, -1, sharpen)
ax[3][0].set_title("Sharpen image")
ax[3][0].imshow(filtered_image, cmap="gray")

filtered_image = cv2.filter2D(gray, -1, laplacian)
ax[3][1].set_title("Laplacian image")
ax[3][1].imshow(filtered_image, cmap="gray")

filtered_image = cv2.filter2D(gray, -1, laplacian2)
ax[4][0].set_title("Laplacian2 image")
ax[4][0].imshow(filtered_image, cmap="gray")

filtered_image = cv2.filter2D(gray, -1, smallBlur)
ax[4][1].set_title("Small blur image")
ax[4][1].imshow(filtered_image, cmap="gray")
```
